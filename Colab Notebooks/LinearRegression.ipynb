```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

data = {
    'SquareFootage': [1500, 1600, 1700, 1800, 1900],
    'NumBedrooms': [3, 3, 3, 4, 4],
    'HousePrice': [300000, 320000, 340000, 360000, 380000]
}

df = pd.DataFrame(data)

X = df[['SquareFootage', 'NumBedrooms']]
Y = df['HousePrice']

# Ensure you have at least two samples in your test set
if len(df) >= 2:
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
    model = LinearRegression().fit(X_train, Y_train)
    Y_pred = model.predict(X_test)

    # Evaluate the model
    r2 = r2_score(Y_test, Y_pred)
    print(f'R²: {r2}')
else:
    print("Cannot calculate R²: Not enough samples in the test set.")
```
