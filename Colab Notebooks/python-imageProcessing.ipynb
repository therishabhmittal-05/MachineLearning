```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
filePath = '/content/drive/MyDrive/glen-carrie-XfA12IEoUkc-unsplash.jpg'
image = cv.imread(filePath, cv.IMREAD_GRAYSCALE)
f_transform = np.fft.fft2(image)
f_transform_shifted = np.fft.fftshift(f_transform)

# Compute the magnitude spectrum
magnitude_spectrum = 20 * np.log(np.abs(f_transform_shifted))

# Plot the original image and its magnitude spectrum
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.title('Magnitude Spectrum')
plt.imshow(magnitude_spectrum, cmap='gray')

plt.show()
print(image)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
filePath = '/content/drive/MyDrive/glen-carrie-XfA12IEoUkc-unsplash.jpg'

image = cv2.imread(filePath, cv2.IMREAD_GRAYSCALE)

equalized_image = cv2.equalizeHist(image)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')

plt.show()

# Display the histograms
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Histogram of Original Image')
plt.hist(image.ravel(), bins=256, range=(0, 256), density=True)

plt.subplot(1, 2, 2)
plt.title('Histogram of Equalized Image')
plt.hist(equalized_image.ravel(), bins=256, range=(0, 256), density=True)

plt.show()
```
