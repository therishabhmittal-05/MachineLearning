```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/kaggle/input/social-network-ads/Social_Network_Ads.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

```python
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# X_train = sc.fit_transform(X_train)
# X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=5, weights = 'distance')
neigh.fit(X_train, y_train)
```

```python
y_predict = neigh.predict(X_test)
```

```python
result = np.concatenate((y_predict.reshape(len(y_predict),1),y_test.reshape(len(y_test),1)),1)
result
```

```python
# Metrics no parameters changed, without scaling
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
print(accuracy_score(y_test, y_predict))
print(precision_score(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
tn, fp, fn, tp = confusion_matrix(y_test, y_predict).ravel()
print(tn, fp, fn, tp)
```

```python
# With feature scaling, change in accuracy and precision_score
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
print(accuracy_score(y_test, y_predict))
print(precision_score(y_test, y_predict))
print(confusion_matrix(y_test, y_predict))
tn, fp, fn, tp = confusion_matrix(y_test, y_predict).ravel()
print(tn, fp, fn, tp)
```
